```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 10))
sns.heatmap(df.iloc[:, :27].corr(), cmap='coolwarm', annot=False, cbar=True)
plt.title("Macierz korelacji cech (atrybuty 1-27)")
plt.show()
```

```python
# targets = df.iloc[:, 27:] # Wybieramy ostatnie 7 kolumnabs
targets = Y
class_counts = targets.sum().sort_values(ascending=False)
print(targets.sum())

plt.figure(figsize=(10, 6))
class_counts.plot(kind='bar')
plt.title("Liczba wystąpień poszczególnych wad")
plt.ylabel("Liczba rekordów")
plt.show()
```
